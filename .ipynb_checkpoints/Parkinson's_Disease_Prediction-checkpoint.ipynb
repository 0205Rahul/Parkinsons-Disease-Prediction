```python
X = df.drop(columns=['class'],axis=1)
Y=df['class']
```

```python
Y
```

```python
print(X)
```

```python
# to split the data in test and training and in the ratio 80:20 
# 80 traiing data and 20 test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# to see the number of training data
print(X.shape, X_train.shape, X_test.shape)
```

```python
ss= StandardScaler()
```

```python
ss.fit(X_train)
```

```python
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# training the SVM with test data
model = svm.SVC(kernel = 'rbf')
```

```python
# training SVm with test data.
model.fit(X_train, Y_train)
```

```python
# model evaluation 
## checking the model accuracy on the training dataset
X_train_pred = model.predict(X_train)
train_data_acc = accuracy_score(Y_train, X_train_pred)
```

```python
# checking the model accuracy on the training dataset
print('acc of training data:', train_data_acc)
```

```python
# checking the model accuracy on the test datset dataset
X_test_pred = model.predict(X_test)
test_data_acc = accuracy_score(Y_test, X_test_pred)
```

```python
print("acc of testing data :", test_data_acc)
```
